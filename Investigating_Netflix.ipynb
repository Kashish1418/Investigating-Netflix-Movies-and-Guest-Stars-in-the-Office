## Subsetting the Data

```python
# Subsetting the dataframe to isolate movies released in the 1990s.
netflix_df['release_year'].value_counts()
```

```python
subset = netflix_df['release_year'].between(1990,1999)
netflix_df = netflix_df[subset]
netflix_df['release_year'].value_counts()
```

```python
# Creating a histogram to visualize the count of movies released in the 1990s
plt.figure(figsize=(10, 6))
plt.hist(netflix_df['release_year'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel("Year")
plt.ylabel("Count of Movies")
plt.title("Distribution of Movies Released in the 90s")
plt.grid(True)
plt.show()
```

--> Netflix featured a higher proportion of movies premiered in the late 1990s than those released in the early amid mid 1990s.


# **Question 1:**
## What was the most frequent movie duration in the 1990s? Save an approximate answer as an integer called duration.

```python
netflix_df['duration'].value_counts()
```

```python
# Using the "mode" in built statistics function to deduce the most frequent movies.
duration = netflix_df['duration'].mode().sum()
duration
```

--> The most frequent movie duration in the 1990s is 94 minutes.

```python
# Comfirming the data type of the duration is in integer or not
duration.dtype
```

**Answer:** On Netflix, the most frquent movies duration in the 1990s is 94 minutes as an integer.

# Question 2:
## A movie is considered short if it's less than 90 minutes. Count the number of short action movies released in the 1990s and save the integer as short_movie_count.

```python
# Checking out the different genre of movies released in the 1990s
netflix_df['genre'].value_counts()
```

```python
# Creating a bar plot to visualize the genre of movies in the 1990s
netflix_df['genre'].value_counts().plot(kind='bar', figsize= (16,8), rot=0)
plt.xlabel("Genre")
plt.ylabel("Counts of Movies")
plt.title("Genre of Movies released in the 1990s")
plt.show();
```

## Subsetting the short action movies with duration less than 90 minutes.

```python
short_action_movies = (netflix_df['genre']== 'Action') & (netflix_df['duration'] < 90)
short_action_movies
```

```python
# Obtaining the total count of movies with short duration
total_short_movies = short_action_movies.sum()
total_short_movies
```

**Answer:** On Netflix, there are only 7 short action movies with duration less than 90 minutes. They were premiered in the 1990s.
